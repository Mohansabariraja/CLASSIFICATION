```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv("/content/diabetes.csv")
```

```python
d.head()
```

```python
d.isnull().sum()
```

```python
x = d.drop("Outcome",axis = 1)
y = d["Outcome"]
```

```python
x_test , x_train , y_test , y_train = train_test_split(x,y,test_size = 0.2,random_state=2)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
accuracy = model.score(x_test,y_test)
print(accuracy)
```

```python
y_pred = model.predict(x_test)
cf = classification_report(y_test,y_pred)
print(cf)
```

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
from sklearn import tree
tree.plot_tree(model, filled = True)
plt.show()
```

```python
graphviz.Source(tree.export_graphviz(model, out_file=None, feature_names = x.columns, filled = True))
```
